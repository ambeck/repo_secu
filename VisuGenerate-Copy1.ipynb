```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_= pd.read_csv("metrics.csv")

df_
```

```python
df_DT = df_[df_['Model_Type'] == "<class 'sklearn.tree.tree.DecisionTreeClassifier'>"]
df_kNN = df_[df_['Model_Type'] == "<class 'sklearn.neighbors.classification.KNeighborsClassifier'>"]
df_NB = df_[df_['Model_Type'] == "<class 'sklearn.naive_bayes.BernoulliNB'>"]
df_aNN = df_[df_['Model_Type'] == "<class 'sklearn.neural_network.multilayer_perceptron.MLPClassifier'>"]
df_RF = df_[df_['Model_Type'] == "<class 'sklearn.ensemble.forest.RandomForestClassifier'>"] 
```

```python
df_DT_bio = (df_DT[df_DT['Train_Set']== 'bio'])
df__DT_astro = df_DT[df_DT['Train_Set'] == 'astro']
df_DT_chem = df_DT[df_DT['Train_Set'] == 'chem']

df_kNN_bio = (df_kNN[df_kNN['Train_Set']== 'bio'])
df_kNN_astro = df_kNN[df_kNN['Train_Set'] == 'astro']
df_kNN_chem = df_kNN[df_kNN['Train_Set'] == 'chem']

df_NB_bio = (df_NB[df_NB['Train_Set']== 'bio'])
df_NB_astro = df_NB[df_NB['Train_Set'] == 'astro']
df_NB_chem = df_NB[df_NB['Train_Set'] == 'chem']

df_aNN_bio = (df_aNN[df_aNN['Train_Set']== 'bio'])
df_aNN_astro = df_aNN[df_aNN['Train_Set'] == 'astro']
df_aNN_chem = df_aNN[df_aNN['Train_Set'] == 'chem']

df_RF_bio = (df_RF[df_RF['Train_Set']== 'bio'])
df_RF_astro = df_RF[df_RF['Train_Set'] == 'astro']
df_RF_chem = df_RF[df_RF['Train_Set'] == 'chem']
```

```python
labels = []
for index, row in df_DT.iterrows():
    labels.append('{}->{}'.format(row['Train_Set'], row['Target_Set']))
print(labels)

p_DT = df_DT[['Insecure_Precision_SMOTE']].to_numpy().flatten()
r_DT = df_DT[['Insecure_Recall_SMOTE']].to_numpy().flatten()
f_DT = df_DT[['Insecure_F_SMOTE']].to_numpy().flatten()

#print(p)

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(30,10))

rects2 = ax.bar(x + width/2, r_DT, width, label='Recall')
rects3 = ax.bar(x + width, f_DT, width, label='F1')
rects1 = ax.bar(x , p_DT, width, label='Precision')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by Train -> Test Pair for Decision Tree')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

fig.tight_layout()

plt.xticks(rotation=45)
plt.show()
plt.savefig('DTmetrics.png')
```
